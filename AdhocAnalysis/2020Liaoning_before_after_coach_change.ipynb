```python
whole_season = df
before_change = df.loc[df['Game_ID']<'2019202000343'].copy()
after_change = df.loc[df['Game_ID']>='2019202000343'].copy()
```

# 整个赛季所有球队数据

### 综合整个赛季来看，广东新疆辽宁三强net rating和新浪积分榜排名相符。
新浪积分榜 http://cba.sports.sina.com.cn/cba/stats/teamrank/?dpc=1

```python
# 整个赛季数据
teams = Team('',whole_season)

teams_whole_season = pd.concat([teams.mov, teams.tm_pace, teams.tm_ortg, teams.tm_drtg, teams.tm_nrtg], axis=1)
teams_whole_season.columns=['场均净胜分MOV', 'Pace', 'OffensiveRating', 'DefensiveRating', 'NetRating']
teams_whole_season = teams_whole_season.sort_values(by='NetRating',ascending=False)
teams_whole_season.rename(index={'辽宁':"辽宁（整赛季）"},inplace=True)
print('本赛季辽宁共打了'+str(teams.tm_ngames['辽宁'])+"场比赛")
teams_whole_season.round(1)
```

# 换帅前数据

### 然而，假如我们只截取辽宁换帅前的比赛数据（换帅前，辽宁共打了34场比赛），辽宁的净效率排名只有第7。

```python
teams = Team('',before_change)

teams_before_change = pd.concat([teams.mov, teams.tm_pace, teams.tm_ortg, teams.tm_drtg, teams.tm_nrtg], axis=1)
teams_before_change.columns=['场均净胜分MOV', 'Pace', 'OffensiveRating', 'DefensiveRating', 'NetRating']
teams_before_change = teams_before_change.sort_values(by='NetRating',ascending=False)
teams_before_change.rename(index={'辽宁':"辽宁（换帅前）"},inplace=True)
print('换帅前辽宁共打了'+str(teams.tm_ngames['辽宁'])+"场比赛")
teams_before_change.round(1)
```

# 换帅后数据
### 假如我们只截取换帅后的数据（换帅后辽宁共打了13场比赛），辽宁的净效率排到联盟第二。

```python
teams = Team('',after_change)

teams_after_change = pd.concat([teams.mov, teams.tm_pace, teams.tm_ortg, teams.tm_drtg, teams.tm_nrtg], axis=1)
teams_after_change.columns=['场均净胜分MOV', 'Pace', 'OffensiveRating', 'DefensiveRating', 'NetRating']
teams_after_change = teams_after_change.sort_values(by='NetRating',ascending=False)
teams_after_change.rename(index={'辽宁':"辽宁（换帅后）"},inplace=True)
print('换帅后辽宁共打了'+str(teams.tm_ngames['辽宁'])+"场比赛")
teams_after_change.round(1)
```

# 整合数据
### 假如我们将换帅后的辽宁，换帅前的辽宁，与全联盟整赛季的球队高阶数据进行对比（包括辽宁队本身的整赛季数据），可以看到辽宁换帅后的净效率高居全联盟第一。当然，也跟换帅后的赛程有关（换帅后并未与目前联盟积分前四的广东，新疆，北京交手）。

```python
merged_df = pd.concat([teams_whole_season.round(1),
                       pd.DataFrame(teams_before_change.loc['辽宁（换帅前）',:].round(1)).transpose(),
                       pd.DataFrame(teams_after_change.loc['辽宁（换帅后）',:].round(1)).transpose()])
merged_df = merged_df.sort_values(by='NetRating',ascending=False)
merged_df
```
