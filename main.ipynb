### To run this Jupyter Notebook, you will need 
1. cbastats folder, which should contain DBHelper.py. Download here https://github.com/JJnotJimmyJohn/CBA-stats/tree/master/cbastats
2. a ".env" file. Content should look like below. Replace password, username with your own username and password. Note: Don't include " " before or after "=".

MONGODB_PWD=password

MONGODB_USERNAME=username

MONGODB_ENDPOINT=cbastats.wllis.mongodb.net/


```python
from cbastats import DBHelper
import datetime
# import pytz
import pandas as pd
```

```python
mongodbio =DBHelper.MongoDBHelper() 
client = mongodbio.create_connection(DBHelper.MONGODB_USERNAME,DBHelper.MONGODB_PWD,DBHelper.MONGODB_ENDPOINT)

db = client['cbaStats']
```

```python
db.list_collection_names()
```

# 球队高阶数据统计

```python
team_stats = pd.DataFrame(mongodbio.select_records(db['seasonTeamStats']),)
team_stats.drop('_id',axis=1,inplace=True)
team_stats
```

# 球员高阶数据

```python
player_stats = pd.DataFrame(mongodbio.select_records(db['seasonPlayerStats']),)
player_stats.drop('_id',axis=1,inplace=True)
player_stats.head()
```
