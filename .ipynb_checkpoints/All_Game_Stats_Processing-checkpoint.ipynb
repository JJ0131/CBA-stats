```python
import pandas as pd
import numpy as np
import re
import xpinyin
```

```python
games_stats = pd.read_csv('StatsData/All_Games_Stats_2020-01-01.csv',encoding='UTF-8',dtype={'Game_ID':object})
# games_stats.drop(axis=1,columns='Unnamed: 0',inplace=True)
games_stats.head()
```

```python
games_stats.loc[games_stats['球员']=='10','球员']= '田宇恒'
```

```python
games_stats.info()
```

```python
player_stats = games_stats.loc[games_stats['球员']=='夏钰博',:]
filter_df = player_stats[['Game_ID','对手']].copy().drop_duplicates()
```

```python
filter_df
```

```python
tm_stats = games_stats.merge(filter_df,left_on=['Game_ID','球队'],right_on=['Game_ID','球队'])
tm_stats.sum(numeric_only=True)
```

# Team Stats

```python
teams_stats = games_stats.groupby(['球队']).sum()
# teams_games = games_stats.groupby(['球队'])['Game_ID'].nunique()

# teams_stats = teams_stats.merge(teams_games,left_index=True,right_index=True)
teams_stats.reset_index(inplace = True)
teams_stats['场次'] = teams_stats['首发']/5
# teams_stats.rename(columns = {'Game_ID':'场次'})
teams_stats
```

```python
teams_stats.shape
```

# Opponents Stats

```python
opponents_stats = games_stats.groupby(['对手']).sum()
opponents_games = games_stats.groupby(['对手'])['Game_ID'].nunique()

opponents_stats = opponents_stats.merge(opponents_games,left_index=True,right_index=True)
opponents_stats.reset_index(inplace = True)
# opponents_stats['场次'] = teams_stats['首发']/5
opponents_stats.rename(columns = {'Game_ID':'场次'},inplace=True)
opponents_stats.rename(columns = {'对手':'对手归属'},inplace=True)
opponents_stats
```

```python
opponents_stats.shape
```

# Player Stats

```python
players_stats = games_stats.groupby(['球员','球队']).sum()
players_stats
```

```python
players_games = games_stats.groupby(['球员'])['Game_ID'].nunique()
players_stats = players_stats.merge(players_games,left_index=True,right_index=True)
players_stats.reset_index(inplace = True)
players_stats.rename(columns = {'Game_ID':'场次'},inplace=True)
players_stats
```
